```python
# The purpose of gradient descent is to find the minimum value of a function by iteratively changing its parameters. 
# It's an optimization algorithm used in machine learning to train models and neural networks. 
# Gradient descent minimizes the errors between predicted and actual results.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x): 
    return (x + 3)**2
```

```python
def derivative(x):
    return 2*(x+3)
```

```python
def gradient_descent(starting_pt, step_size, iteration):
    x = starting_pt
    points = []
    
    for i in range(iteration):
        points.append(x)
        print("iteration", i+1)
        print("x = ", x)
        gradient = derivative(x);
        x = x - gradient * step_size
        
    return x, points
```

```python
starting_pt = 2
step_size = 0.1 
iteration = 1000

minimum, points = gradient_descent(starting_pt, step_size, iteration)
print("minimum = ", minimum)
```

```python
x  = np.linspace(-5, 5, 100)
y = function(x)

plt.plot(x, y, label = "y = (x+3)^2")
# display all the points with red dot
plt.scatter(points, [function(x) for x in points], color="red", label = "red")
plt.grid()
plt.title("Graph")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
